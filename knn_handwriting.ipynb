```python
# -*- coding:utf-8 -*-
import numpy as np
import operator
import os
```

```python
'''
将一张图片的像素点值txt转化为数组，规格为32x32
'''
def img2vector(filename):
    returnVect = np.zeros((1,1024))# 数组规格1x1024
    fr = open(filename) 
    for i in range(0,32): 
        lineStr = fr.readline() # 读取每行像素值，存储进数组
        for j in range(0,32):
            returnVect[0,32*i+j] = int(lineStr[j])
    return returnVect 
```

```python
'''
实现kNN分类
'''
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat**2 
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5 # 计算待分类样本与已有数据集样本点的L2范数，即欧氏距离
    sortedDistIndicies = distances.argsort() # 将待分类样本与有监督样本间的距离排序，提取有监督样本的对应索引
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]] # 提取距离最小的前k个样本的类别标签
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)# 将前k个小距离样本中不同类别出现次数降序排列
    return sortedClassCount[0][0] # 返回出现次数最多的类为待分类样本的预测标签
```

```python
def handwritingClassTest():
    hwLabels = []
    trainingFileList = os.listdir('trainingDigits')
    m = len(trainingFileList) # 训练集样本个数为m
    trainingMat = np.zeros((m,1024)) # trainingMat按行存储训练集中所有像素点，总规模m*1024
    for i in range(0,m): # 每个txt文件名格式为：数字类别_文件编号.txt
        fileNameStr = trainingFileList[i] 
        fileStr = fileNameStr.split('.')[0] # 提取文件名，即.txt前面的部分
        classNumStr = int(fileStr.split('_')[0]) # 提取本张手写数字的类标签
        hwLabels.append(classNumStr) # hwLabel列表用于存储训练集中所有张手写数字的类标签
        trainingMat[i,:] = img2vector('trainingDigits/'+ fileNameStr) # 将第i个手写数字文件的像素值存储进trainingMat的第i行
    testFileList = os.listdir('testDigits')# 测试集
    errorCount = 0 # 错误分类数，初始化为0
    mTest = len(testFileList) # 测试集样本个数为mTest
    for i in range(0,mTest): # 开始对测试集每个样本类别进行预测
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0]) # 提取其真实类标签
        vectorUnderTest = img2vector('testDigits/'+ fileNameStr) # 第i个手写数字文件的像素值存储进vectorUnderTest数组
        classifierResult = classify0(vectorUnderTest,trainingMat,hwLabels,3) # 输入待分类样本特征，训练集样本，训练集标签，kNN中k值
        print('the classifier came back with: %d, the real answer is: %d' % (classifierResult,classNumStr)) # 输出当前待分类样本的预测类标签
        if classifierResult != classNumStr: # 判断是否错误分类，统计错分样本数
            errorCount += 1
        print('\nthe total number of errors is: %d' % errorCount)# 输出错分样本总数
        print('\nthe total error rate is: %f' % (errorCount/float(mTest))) # 计算分类错误率并输出
    return
```

```python
handwritingClassTest() # 执行手写数字分类
```
